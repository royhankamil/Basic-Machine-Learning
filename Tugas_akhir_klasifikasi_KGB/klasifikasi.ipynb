```python
base_dir = '/tmp/rockpaperscissors'
paper_dir = base_dir + '/paper'
rock_dir = base_dir + '/rock'
scissors_dir = base_dir + '/scissors'
```

pengumpulan path datasets melalui

```python
rock = [os.path.join(rock_dir, file) for file in os.listdir(rock_dir)]
paper = [os.path.join(paper_dir, file) for file in os.listdir(paper_dir)]
scissors = [os.path.join(scissors_dir, file) for file in os.listdir(scissors_dir)]

tot_datasets = rock + paper + scissors

train_set, val_set = train_test_split(tot_datasets, test_size=0.4, random_state=42)
```

memindahkan datasets dan melakukan pemisahan folder melalui label

```python
new_path = "/content/datasets"
train_path = new_path + "/train"
val_path = new_path + "/validation"
category = ['scissors', 'rock', 'paper']

def labeling(path):
  for element in category:
    os.makedirs(path + "/" + element, exist_ok=True)

os.makedirs(new_path, exist_ok=True)
os.makedirs(train_path, exist_ok=True)
os.makedirs(val_path, exist_ok=True)

labeling(train_path)
labeling(val_path)

try:
  for path in train_set:
    if "/tmp/rockpaperscissors/rock" in path:
      shutil.move(path,train_path + "/rock")
    if "/tmp/rockpaperscissors/paper" in path:
      shutil.move(path,train_path + "/paper")
    if "/tmp/rockpaperscissors/scissors" in path:
      shutil.move(path,train_path + "/scissors")

  for path in val_set:
    if "/tmp/rockpaperscissors/rock" in path:
      shutil.move(path,val_path + "/rock")
    if "/tmp/rockpaperscissors/paper" in path:
      shutil.move(path,val_path + "/paper")
    if "/tmp/rockpaperscissors/scissors" in path:
      shutil.move(path,val_path + "/scissors")

except shutil.Error:
  print("already moved")

```

proses augmentasi gambar

```python
train_datagen = ImageDataGenerator(
                rescale=1./255,
                rotation_range=20,
                horizontal_flip=True,
                shear_range=0.2,
                fill_mode='nearest'
)

test_datagen = ImageDataGenerator(
                rescale=1./255
)
```

```python
train_generator = train_datagen.flow_from_directory(
        train_path,
        target_size=(200, 300),
        batch_size=4,
        class_mode='categorical')

validation_generator = test_datagen.flow_from_directory(
        val_path,
        target_size=(200, 300),
        batch_size=4,
        class_mode='categorical')
```

membangun model CNN

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(200, 300, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.4),

    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.4),

    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.4),

    tf.keras.layers.Conv2D(512, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.4),

    tf.keras.layers.GlobalAveragePooling2D(),

    tf.keras.layers.Dense(512, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)),
    tf.keras.layers.Dropout(0.3),

    tf.keras.layers.Dense(3, activation='softmax')
])
```

```python
model.summary()
```

melakukan compiling pada model dan training model

```python
model.compile(loss='categorical_crossentropy', optimizer=tf.optimizers.Adam(learning_rate=0.001), metrics=['accuracy'])
```

```python
model.fit(
    train_generator,
    steps_per_epoch=30,
    epochs=25,
    validation_data=validation_generator,
    validation_steps=5,
    verbose=2
    )
```

```python
train_loss, train_accuracy = model.evaluate(train_generator)

val_loss, val_accuracy = model.evaluate(validation_generator)

print(f"Training Accuracy: {train_accuracy*100:.2f}%")
print(f"Validation Accuracy: {val_accuracy*100:.2f}%")
```

```python
import numpy as np
from google.colab import files
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline

uploaded = files.upload()
labels = {0: 'paper', 1: 'rock', 2: 'scissors'}

for fn in uploaded.keys():
    path = fn
    img = image.load_img(path, target_size=(200, 300))

    imgplot = plt.imshow(img)
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    images = np.vstack([x])

    classes = model.predict(images, batch_size=4)
    print(classes)
    result = labels[classes.argmax()]
    print(result)
    if result=='paper':
        print('paper')
    elif result=='rock':
        print('rock')
    elif result=='scissors':
        print('scissors')
```
